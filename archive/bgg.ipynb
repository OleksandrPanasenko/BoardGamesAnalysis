```python
def properties_to_tables(property_table, identifier):
    """
    Converts a binary property DataFrame into an intermediary SQL-style table.

    :param property_table: pandas DataFrame with binary indicators (1/0 or True/False)
    :param identifier: list or Series of object identifiers (same length as property_table)
    :return: (list of (object_id, property_id) pairs, dict of {property_id: property_name})
    """
    # Step 1: Map properties to integer IDs
    features = {i + 1: col for i, col in enumerate(property_table.columns)}
    property_to_id = {v: k for k, v in features.items()}

    # Step 2: Add the object identifier as a column
    df = property_table.copy()
    df['object_id'] = identifier

    # Step 3: Melt the table into long format
    df_long = df.melt(id_vars='object_id', var_name='property', value_name='has_property')

    # Step 4: Filter only rows where the property is set (1 or True)
    df_filtered = df_long[df_long['has_property'] == 1]

    # Step 5: Replace property name with property_id
    df_filtered['property_id'] = df_filtered['property'].map(property_to_id)

    # Step 6: Extract only (object_id, property_id)
    relations = list(df_filtered[['object_id', 'property_id']].itertuples(index=False, name=None))

    relations=pd.DataFrame(relations,index=range(0,len(relations)))

    return relations, features
```
